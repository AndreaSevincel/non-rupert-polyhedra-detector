# Data Exploration of Convex Polyhedra

In this notebook, we will explore the dataset of convex polyhedra. We will perform statistical analysis and visualize the properties of the polyhedra to gain insights into their characteristics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid')
```

```python
# Load the dataset
data_path = '../data/processed/polyhedra_dataset.csv'
polyhedra_df = pd.read_csv(data_path)

# Display the first few rows of the dataset
polyhedra_df.head()
```

```python
# Summary statistics of the dataset
polyhedra_df.describe()
```

```python
# Visualize the distribution of a specific property (e.g., volume)
plt.figure(figsize=(10, 6))
sns.histplot(polyhedra_df['volume'], bins=30, kde=True)
plt.title('Distribution of Polyhedra Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap of the properties
plt.figure(figsize=(12, 8))
correlation_matrix = polyhedra_df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Polyhedra Properties')
plt.show()
```
